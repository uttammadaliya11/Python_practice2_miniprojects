```python
import numpy as np
from scipy.stats import norm
```

```python
Incomes=[157434,261952,144657,118777,147511,163025,120131,162745,155569]
Education=[0,2,0,1,1,2,0,0,1]
```

```python
mu1=188014.34
mu2=143104.34
mu3=146148.34
sigma1=52541.39
sigma2=18330.95
sigma3=18628.83
```

```python
f=[]

for income in Incomes:
    tmp=[]
    tmp.append(norm.pdf(income, loc=mu1, scale=sigma1))
    tmp.append(norm.pdf(income, loc=mu2, scale=sigma2))
    tmp.append(norm.pdf(income, loc=mu3, scale=sigma3))
    f.append(tmp)


for i in range (len(f)):
    f[i]=f[i]/sum(f[i])
```

```python
f_t=np.transpose(f)
```

```python
#find the new means:

tmp1=0
tmp2=0
tmp3=0
for i in range (len(f)):
    tmp1+=f[i][0]*Incomes[i]
    tmp2+=f[i][1]*Incomes[i]
    tmp3+=f[i][2]*Incomes[i]


mu1=tmp1/sum(f_t[0])
mu2=tmp2/sum(f_t[1])
mu3=tmp3/sum(f_t[2])
print("mu1:",mu1)
print("mu2:",mu2)
print("mu3:",mu3)
```

```python
#find the new stds:

tmp1=0
tmp2=0
tmp3=0
for i in range (len(f)):
    tmp1+=f[i][0]*pow((Incomes[i]-mu1),2)
    tmp2+=f[i][1]*pow((Incomes[i]-mu2),2)
    tmp3+=f[i][2]*pow((Incomes[i]-mu3),2)


sigma1=np.sqrt(tmp1/sum(f_t[0]))
sigma2=np.sqrt(tmp2/sum(f_t[1]))
sigma3=np.sqrt(tmp3/sum(f_t[2]))
```

```python
print("sigma1:",sigma1)
print("sigma2:",sigma2)
print("sigma3:",sigma3)
```
